```python
Identity = pd.read_csv('/kaggle/input/test_identity.csv')
Transaction = pd.read_csv('/kaggle/input/test_transaction.csv')
Identity.columns = Identity.columns.str.replace('-', '_')
```

```python
df = Transaction.merge(Identity, how='left', left_on='TransactionID', right_on='TransactionID')
```

```python
import mlflow
mlflow.set_tracking_uri("https://dagshub.com/Givi-Modebadze/Fraud_Detection_ML.mlflow")

loaded_model = mlflow.sklearn.load_model('runs:/51704e8765134ee5bfc3a533bfda57ca/xgboost_pipeline_model')
```

```python
X = df

probas = loaded_model.predict_proba(X)[:, 1]

submission_df = pd.DataFrame({
    'TransactionID': X['TransactionID'],
    'isFraud': probas
})

submission_df.to_csv("submission.csv", index=False)

submission_df
```
